```python
# Iport Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Storing Dataset
dataset=pd.read_csv("Placement_Data_Full_Class.csv")
dataset
```

```python
import seaborn as sb
```

```python
# For Categorcal Plot
```

```python
sb.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)
plt.show()
```

```python
#Create swarmplot for SSLC Pass Mark
sb.swarmplot(y="mba_p",data=dataset)
plt.show()
```

```python
# Create swarmplot group by categories 
sb.swarmplot(x="specialisation",y="mba_p",data=dataset)
plt.show()
```

```python
# Create swarmplot group by categories using hue
sb.swarmplot(x="specialisation",y="mba_p",data=dataset,hue="gender")
plt.show()
```

```python
# Create swarmplot showing Hue seperately on the categories axis
sb.swarmplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True)
plt.show()
```

```python
# Create swarmplot showing shape and size
plt.figure(figsize=(15,10))
sb.swarmplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True, marker="*",size=10,palette="dark:blue",edgecolor="green")
plt.show()
```

```python
# Overlay swarmplot on Box plot
sb.boxplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True)
sb.swarmplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True)
plt.show()
```

```python
# Overlay swarmplot on violin plot
sb.violinplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True)
sb.swarmplot(x="specialisation",y="mba_p",data=dataset,hue="gender",dodge=True)
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset)
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```

```python
sb.violinplot(x="specialisation",y="salary",data=dataset,hue="gender",split=True)
plt.show()
```

```python
sb.violinplot(x="gender",y="ssc_p",data=dataset)
plt.show()
```

```python
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender")
plt.show()
```

```python
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender",split=True)
plt.show()
```

```python
# Change the box in the violin plot o horizonal lines
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender",split=True,inner="quartile")
plt.show()
```

```python
# Drwa line for each observaton in violin plot
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender",split=True,inner="stick")
plt.show()
```

```python
# Change he amount of smoothing using bw attribue
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender",split=True,inner="quartile",bw_method=.2)
plt.show()
```

```python
# cut out te extream values
sb.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender",split=True,inner="quartile",bw_method=.2, cut=0)
plt.show()
```
